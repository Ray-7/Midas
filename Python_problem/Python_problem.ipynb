# Getting URL of the user

---



```
def getURL(screen_name):
    URL = 'https://api.twitter.com/1.1/statuses/user_timeline.json?screen_name={}'.format(screen_name)
    return URL
```

# Getting number of tweets from the user

---



```
username = "midasIIITD"
URL = getURL(username)
tweets_request = requests.get(URL, auth=auth)
tweet_count = tweets_request.json()[0]["user"]["statuses_count"]
```

# Getting tweets

---



```
URL = URL + '&tweet_mode=extended&count={}'.format(tweet_count)
tweets_request = requests.get(URL, auth=auth)
tweets = tweets_request.json()
```

# Counting the number of images in a tweet

---



```
def images_count(tweet):
    count = 0
    if 'media' in tweet['entities']:
        count = len(tweet['entities']['media'])
    if 'extended_entities' in tweet:
        if 'media' in tweet['extended_entities']:
            count = count + len(tweet['entities']['media'])
    return count
```

# Representation of Data in a Tabular format

---



```
col_labels = ["Date/Time","Full_Text","Favorites","Retweets","Image_Count"]
```

```
pd.set_option('display.max_colwidth', -1)
tabular_data = pd.DataFrame(columns = col_labels)
for i, tweet in enumerate(tweets):
    
    image_count = images_count(tweet)
    if image_count == 0:
        count = 'None'
    else:
        count = str(image_count)
    
    tabular_data.loc[i] = [
        tweet['created_at'],
        tweet['full_text'],
        tweet['favorite_count'],
        tweet['retweet_count'],
        count
                          ]

    
tabular_data
```
